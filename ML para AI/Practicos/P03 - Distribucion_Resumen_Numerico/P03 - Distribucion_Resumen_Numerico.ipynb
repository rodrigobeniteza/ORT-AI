Esta notebook sirve de apoyo para resolver el ejercicio de estadística cuyo objetivo es bosquejar la distribución de notas de una clase a partir de un resumen numérico

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
N = 100
```

```python
datos = [62]*N + [65] + [94]*N
```

```python
datos
```

```python
np.mean(datos)
```

```python
np.median(datos)
```

```python
np.std(datos)
```

```python
sns.kdeplot(datos,fill=True,bw_adjust=0.125)
plt.show()
```
